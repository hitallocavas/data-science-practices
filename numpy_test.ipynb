# Exercícios NumPy  

Agora que aprendemos sobre NumPy, vamos testar seu conhecimento. Começaremos com algumas tarefas simples, para depois entrarmos nas perguntas mais complicadas.

#### Importe NumPy como np

```python
import numpy as np
```

#### Crie uma matriz de 10 zeros

```python
matriz_10zeros = np.zeros(10)
matriz_10zeros
```

#### Crie uma matriz de 10 ones

```python
matriz_10ones = np.ones(10)
matriz_10ones
```

#### Crie uma matriz de 10 cincos

```python
matriz_10cincos = matriz_10zeros + 5
matriz_10cincos
```

#### Crie um array de inteiros de 10 até 50

```python
array10inteiros_dez_ciquenta = np.arange(10,51)
array10inteiros_dez_ciquenta
```

#### Crie um array dos numeros pares de 10 até 50

```python
arraypares_dez_cinquenta = np.arange(10,51,2)
arraypares_dez_cinquenta
```

#### Criei uma matriz 3x3 com valores variando de 0 até 8

```python
arr_3_3_zero_oito = np.arange(9)
matriz_3_3 = arr_3_3_zero_oito.reshape(3,3)
matriz_3_3
```

#### Crie uma matriz identidade 3x3

```python
matriz_identidade_3_3 = np.eye(3)
matriz_identidade_3_3
```

#### Use NumPy para gerar números aleatórios entre 0 e 1

```python
num_aleatorio_0_1 = np.random.rand(1)
num_aleatorio_0_1
```

#### Use Numpy para gerar um array de 25 números aleatórios tirados de uma distribuição normal.

```python
arr_25_normal = np.random.randn(25)
arr_25_normal
```

#### Crie a seguinte matriz:

```python
matriz_10_10_zero_100_div_100 = np.arange(1,101).reshape(10,10) / 100
matriz_10_10_zero_100_div_100
```

#### Crie um array de tamanho 20 igualmente espaçado entre 0 e 1.

```python
arr_20_zero_um = np.linspace(0,1,20)
arr_20_zero_um
```

## Indexação Numpy e Seleção

Agora você receberá algumas matrizes e será solicitado a replicar as saídas resultantes da matriz:

```python
mat = np.arange(1,26).reshape(5,5)
mat
```

```python
mat[2:,1:]
```

```python
condition = mat == 20
mat[condition][0]
```

```python
mat[:3,1].reshape(3,1)
```

```python
mat[4]
```
